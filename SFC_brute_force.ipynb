```python
class MEC:
    def __init__(self, mname, pcdelay, res, ids):
        self.name = mname
        self.processorDelay = pcdelay
        self.nores = res
        self.id = ids

    def seeContent(self):
            print(f"MEC name : {self.name} Processor Delay : {self.processorDelay} Number of Resources : {self.nores} Id : {self.id}")
```

```python
class SFC:
    def __init__(self,lat):
        self.VNF = []
        self.latency = lat

    def add_VNF(self,vnf):
        self.VNF.append(vnf)
         #print(self.VNF)

    def viewSF(self):
        print(f"{self.VNF}, {self.latency}")
```

```python
combinations=[]
def generate_subsets(out):
    for subset in itertools.product(*out):
        print(list(subset))
        combinations.append(list(subset))
```
